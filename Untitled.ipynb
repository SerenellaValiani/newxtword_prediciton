```python
import sys
#print(sys.path)
#sys.executable
import nltk
#download data from nltk
#first write 'd', then 'all'
nltk.download()
import string
import re
import pickle
import keras
import numpy as np
from tensorflow.keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer
import csv
import pandas as pd
import numpy as np
import tensorflow
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import TensorBoard


#-----------------------------FUNCTIONS-----------------------------

#return a string of the text from a file
def string_from_file(file):
    #list of afile's lines
    lines = []
    lines = file.readlines()
    #string of file's lines
    data = ''
    data = ' '.join(lines)
    #removing special characters
    data = data.replace('\n', '').replace('\r', '').replace('\ufeff', '')
    return data



#return a list of sentences from a string
def sents_from_string(string):
    sents = nltk.regexp_tokenize(string, r'[,\.\:\;\.\?!"]\s*', gaps=True)
    #substitute symbol ’ with a white space
    for s in sents:
        s = s.replace('’', ' ')
        #replace('-', ' ')
    return sents



#return a vocabulary - a list - of the words used in a text
def vocab_from_string(string):
    #list of all the words, with puntuaction and ripetitions
    rip_word = nltk.word_tokenize(string)
    #list of the words used, without puntuanction and ripetitions
    puntuaction = ['.', ',', ':', ';', '"', '’', '``', "''", '(', ')', '[', ']', '_', '-', '#', '\'' ,'!', '?', '&']
    vocab = []
    for i in rip_word:
        #i = i.lower()
        if i[-1] == ".":
            i = i[0:len(i) - 1]
        if i not in vocab and i not in puntuaction:
            vocab.append(i)
    return vocab



#return a dictionary: tuple(len = 5) -> list
#the first element in the tuple is different from 0
def one_to_one(vocab, sentences):
    next_word1 = {}
    for v in vocab:
        value_list = []
        for s in sentences:
            s_list = nltk.regexp_tokenize(s, r'[,.\.\ \.\:\;\—\_\-\'\’\.\?!"]\s*', gaps=True)
            if v in s_list:
                index = s_list.index(v)
                if index < len(s_list) - 1:
                    value_list.append(s_list[index + 1])
        next_word1[(v, '', '', '', '')] = value_list
    #removing keys with empty value
    del_keys = []
    for key, value in next_word1.items():
        if value == []:
            del_keys.append(key)
    for d in del_keys:
        del next_word1[d]
    return next_word1


#the first and the secondo elements in the tuple are different from 0
def two_to_one(vocab, sentences):
    next_word2 = {}                    
    for key in next_word1:
        for v in next_word1[key]:
            value_list = []
            if (key[0], v, '', '', '') in next_word2.keys():
                continue
            for s in sentences:
                s_list = nltk.regexp_tokenize(s, r'[,.\.\.\:\;\—\_\-\'\’\.\?!"]\s*', gaps=True)
                if key[0] in s_list:
                    index = s_list.index(key[0])
                    if index < len(s_list) - 2:
                        if v == s_list[index + 1]:
                            value_list.append(s_list[index + 2])
            next_word2[(key[0], v, '', '', '')] = value_list
    #removing keys with empty value
    del_keys = []
    for key, value in next_word2.items():
        if value == []:
            del_keys.append(key)
    for d in del_keys:
        del next_word2[d]
    return next_word2


#the first three elements in the tuple are different from 0
def three_to_one(vocab, sentences):
    next_word3 = {}
    for key in next_word2:
        for v in next_word2[key]:
            value_list = []
            if (key[0], key[1], v, '', '') in next_word3.keys():
                continue
            for s in sentences:
                s_list = nltk.regexp_tokenize(s, r'[,.\.\.\:\;\—\_\-\'\’\.\?!"]\s*', gaps=True)
                if key[0] in s_list:
                    index = s_list.index(key[0])
                    if index < len(s_list) - 3:
                        if key[1] == s_list[index + 1]:
                            if v == s_list[index + 2]:
                                value_list.append(s_list[index + 3])
            next_word3[(key[0], key[1], v, '', '')] = value_list
    #removing keys with empty value
    del_keys = []
    for key, value in next_word3.items():
        if value == []:
            del_keys.append(key)
    for d in del_keys:
        del next_word3[d]
    return next_word3


#the first four elements in the tuple are different from 0
def four_to_one(vocab, sentences):
    next_word4 = {}
    for key in next_word3:
        for v in next_word3[key]:
            value_list = []
            if (key[0], key[1], key[2], v, '') in next_word4.keys():
                continue
            for s in sentences:
                s_list = nltk.regexp_tokenize(s, r'[,.\.\.\:\;\—\_\-\'\’\.\?!"]\s*', gaps=True)
                if key[0] in s_list:
                    index = s_list.index(key[0])
                    if index < len(s_list) - 4:
                        if key[1] == s_list[index + 1]:
                            if key[2] == s_list[index + 2]:
                                if v == s_list[index + 3]:
                                    value_list.append(s_list[index + 4])
            next_word4[(key[0], key[1], key[2], v, '')] = value_list
    #removing keys with empty value
    del_keys = []
    for key, value in next_word4.items():
        if value == []:
            del_keys.append(key)
    for d in del_keys:
        del next_word4[d]
    return next_word4


#the first five elements in the tuple are different from 0
def five_to_one(vocab, sentences):
    next_word5 = {}
    for key in next_word4:
        for v in next_word4[key]:
            value_list = []
            if (key[0], key[1], key[2], key[3], v) in next_word5.keys():
                continue
            for s in sentences:
                s_list = nltk.regexp_tokenize(s, r'[,.\.\.\:\;\—\_\-\'\’\.\?!"]\s*', gaps=True)
                if key[0] in s_list:
                    index = s_list.index(key[0])
                    if index < len(s_list) - 5:
                        if key[1] == s_list[index + 1]:
                            if key[2] == s_list[index + 2]:
                                if key[3] == s_list[index + 3]:
                                    if v == s_list[index + 4]:
                                        value_list.append(s_list[index + 5])
            next_word5[(key[0], key[1], key[2], key[3], v)] = value_list
    #removing keys with empty value
    del_keys = []
    for key, value in next_word5.items():
        if value == []:
            del_keys.append(key)
    for d in del_keys:
        del next_word5[d]
    return next_word5

#from five dictionaries, return two lists: together they represent a training set
#                                          the first one contains the input, the second the output
def training_set(next_word1, next_word2, next_word3, next_word4, next_word5):
    X = []
    Y = []
    for key, value in next_word1.items():
        for i in range(len(value)):
            #writer.writerow([[vocab.index(key[0]), 0, 0, 0, 0], vocab.index(value[i])])
            X.append([vocab.index(key[0]), 0, 0, 0, 0])
            Y.append(vocab.index(value[i]))
    for key, value in next_word2.items():
        for i in range(len(value)):
            #writer.writerow([[vocab.index(key[0]), vocab.index(key[1]) , 0, 0, 0], 
            #                 vocab.index(value[i])])
            X.append([vocab.index(key[0]), vocab.index(key[1]), 0, 0, 0])
            Y.append(vocab.index(value[i]))
    for key, value in next_word3.items():
        for i in range(len(value)):
            #writer.writerow([[vocab.index(key[0]), vocab.index(key[1]), vocab.index(key[2]), 0 , 0], 
            #                 vocab.index(value[i])])
            X.append([vocab.index(key[0]), vocab.index(key[1]), vocab.index(key[2]), 0, 0])
            Y.append(vocab.index(value[i]))
    for key, value in next_word4.items():
        for i in range(len(value)):
            #writer.writerow([[vocab.index(key[0]), vocab.index(key[1]), vocab.index(key[2]), 
            #                 vocab.index(key[3]), 0], vocab.index(value[i])])
            X.append([vocab.index(key[0]), vocab.index(key[1]), vocab.index(key[2]), vocab.index(key[3]), 0])
            Y.append(vocab.index(value[i]))
    for key, value in next_word5.items():
        for i in range(len(value)):
            #writer.writerow([[vocab.index(key[0]), vocab.index(key[1]), vocab.index(key[2]), 
            #                 vocab.index(key[3]), vocab.index(key[4])], vocab.index(value[i])])
            X.append([vocab.index(key[0]), vocab.index(key[1]), vocab.index(key[2]), vocab.index(key[3]), vocab.index(key[4])])
            Y.append(vocab.index(value[i]))
    return X,Y


print('Functions defined.')
```

```python
#---------------------------PREPROCESSING---------------------------
#open the file
# replaced '—' with ' '
# replaced '“' with '"'
# replaced '”' with '"'
file = open('short_text.txt', 'r')

#string of the content of the file
data = string_from_file(file)

#list of file's sentences
sentences = sents_from_string(data)

#vocabulary - a list - obtained from the file
vocab = vocab_from_string(data)

#tokenizzazione
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

pickle.dump(tokenizer, open('tokenizer.pkl', 'wb'))
#sequence_vocab = tokenizer.texts_to_sequences([string_vocab])[0]
vocab_size = len(tokenizer.word_index) + 1
print("token: {}".format(vocab_size))
print("vocab: {}".format(len(vocab)))


#Five dictionaries: tuple -> list
#                   *tuple, len 5: there are the words from the vocabulary ordered through the sentences
#                                  they represent the words that will be read as inputs
#                   *list, len not fixed: the word that follows the words in the tuple according to the sentences
#                                  it represents the word that will be predicted 
#first dictionary, the tuple contains only one word
next_word1 = one_to_one(vocab, sentences)
#second dictionary, the tuple represent a sentence of two words
next_word2 = two_to_one(vocab, sentences)
#third dictionary, the tuple represent a sentence of three words
next_word3 = three_to_one(vocab, sentences)
#fourth dictionary, the tuple represent a sentence of four words
next_word4 = four_to_one(vocab, sentences)
#fifth dictionary, the tuple represent a sentence of five words
next_word5 = five_to_one(vocab, sentences)
print('Dictionaries defined.')

#define the training set
X, Y = training_set(next_word1, next_word2, next_word3, next_word4, next_word5)
print('Training set defined.')
#data_set = open('dataset.csv', 'w', newline='')
#writer = csv.writer(data_set)
#data_set.close()
```

```python
#------------------------------MODEL------------------------------

#data = pd.read_csv("dataset.csv", names = ["input", "output"])
#X = data['input']
#Y = data['output']
X = np.array(X)
Y = np.array(Y)
Y = to_categorical(Y, num_classes=vocab_size)


model = tensorflow.keras.Sequential()
model.add(tensorflow.keras.layers.Embedding(vocab_size, 5, input_length=1))
model.add(tensorflow.keras.layers.LSTM(1000, return_sequences=True))
model.add(tensorflow.keras.layers.LSTM(1000))
model.add(tensorflow.keras.layers.Dense(1000, activation="relu"))
model.add(tensorflow.keras.layers.Dense(vocab_size, activation="softmax"))


checkpoint = ModelCheckpoint("nextword1.h5", monitor='loss', verbose=1, save_best_only=True, mode='auto')
reduce = ReduceLROnPlateau(monitor='loss', factor=0.2, patience=3, min_lr=0.0001, verbose = 1)
logdir='logsnextword1'
tensorboard_Visualization = TensorBoard(log_dir=logdir)


model.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.Adam(lr=0.001))
model.fit(X, Y, epochs=50, validation_split=0.1, batch_size=64, callbacks=[checkpoint, reduce, tensorboard_Visualization])
```

```python
#!/usr/bin/env python
# coding: utf-8
import keras
from tensorflow.keras.models import load_model
import numpy as np
import pickle

#carico il modello e il tokenizer
model = load_model('nextword1.h5')
tokenizer = pickle.load(open('tokenizer.pkl', 'rb'))

def Predict_Next_Words(model, tokenizer, text):

    list_text = text.split()
    list_sequence = [0, 0, 0, 0, 0 ]
    sequence = tokenizer.text_to_sequence([text])[0]
    split_sequence = sequence.split()
    for i in range(len(split_sequence)):
        list_sequence[i] = split_sequence[i]
    sequence = np.array(list_sequence)
    preds = model.predict_classes(sequence)
    predicted_word = ''
    for key, value in tokenizer.word_index.items():
        if value == preds:
            predicted_word = key
            break
    return predicted_word

while(True):
    text = input("Enter your line, using only lowercase letters: ")
    if text == "exit":
        print("Ending The Program.....")
        break
    else:
        try:
            predicted_word = Predict_Next_Words(model, tokenizer, text)
            print('Predicted word: {}'.format(predicted_word))
        except:
            continue

                       
```
